```python
import matplotlib
matplotlib.use("svg")
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import numpy as np

from PIL import Image

def plot_image(im):
    plt.imshow(np.asarray(im))
    
    
image = Image.open("../Data/test_imgs/Coat_Female_3564.jpg")
print(image.format, image.size, image.mode)
plot_image(image)
```

```python
def crop_and_scale_image(im):
    """ Crops and scales a given image. 
        Args: 
            im (PIL Image) : image object to be cropped and scaled
        Returns: 
            (PIL Image) : cropped and scaled image object
    """
    width,height = im.size
    if width > height:
        diff = width - height
        box = diff/2, 0, width - (diff - diff/2), height
    else:
        diff = height - width
        box = 0, diff/2, width, height - (diff - diff/2)
    im = im.crop(box)
    toSize = 100,100
    im= im.resize(toSize, Image.ANTIALIAS)
    return im

plot_image(crop_and_scale_image(image))
```

```python
import os
dname = "../Data/test_imgs/"
im_paths = np.array([dname+fname for fname in os.listdir(dname) if fname.endswith(".jpg")])
print(len(im_paths))
P = np.random.permutation(len(im_paths))
split = 2000
fnames_tr, fnames_va, fnames_te = im_paths[P[:split]], im_paths[P[split:2*split]], im_paths[P[2*split:]]
```

```python
plt.figure(figsize=(20,20))
for i,fname in enumerate(fnames_tr[:100]):
    plt.subplot(10,10,i+1)
    plt.axis('off')
    plot_image(crop_and_scale_image(Image.open(fname)))
```

```python
def fname_to_network_input(fname):
    """ Creates the input for a VGG network from the filename 
        Args: 
            fname (string) : the filename to be parsed
        Returns: 
            (numpy ndarray) : the array to be passed into the VGG network as a single example
    """
    im = Image.open(fname)
    im = crop_and_scale_image(im)
    if im.mode is not 'RGB':
        im = im.convert('RGB')
    npim = np.asarray(im)
    npim = np.rollaxis(npim, 2)
    return npim

network_input = fname_to_network_input("../Data/test_imgs/Coat_Female_3564.jpg")
print(network_input, type(network_input))
```

```python
def get_label(fname):
    """
    The image filename is in format: label_xxx.jpg, 
    where xxx is the number of this image in such label.
    """
    label = "".join(fname.split('_')[:-1])
    return label

label_dict = {}
idx = 0
for fname in os.listdir("../Data/train_imgs/"):
    label = get_label(fname)
    if label not in label_dict:
        label_dict[label] = idx
        idx += 1
print(label_dict)
```

```python
def load_data(path):
    dname = path
    im_paths = []
    labels = []
    for fname in os.listdir(dname):
        if fname.endswith('.jpg'):
            label = get_label(fname)
            if label not in label_dict:
                print('wrong label ', label)
            im_paths.append(dname+fname)
            try:
                labels.append(label_dict[label])
            except:
                continue 
    data = []
    for file in im_paths:
        data.append(fname_to_network_input(file))
        
    data = np.asarray(data)
    labels = np.asarray(labels)
    return data, labels

train_data, train_labels = load_data("../Data/train_imgs/")
test_data, test_labels = load_data("../Data/test_imgs/")
print(train_data.shape, train_labels.shape)
print(test_data.shape, test_labels.shape)
```

```python
for key in label_dict:
    idx = label_dict[key]
    print('{} : {}'.format(idx, sum(train_labels == idx)))
    
for key in label_dict:
    idx = label_dict[key]
    print('{} : {}'.format(idx, sum(test_labels == idx)))
```
